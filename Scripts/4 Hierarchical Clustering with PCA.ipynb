## 1. Import Libraries and Data

```python
# Import Libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#Create a path to where your data is stored.
path = r'/Users/sydneyjohnson/Documents/CF Data Analytics Course/11-2024 ClimateWins Analysis/02 Data'
```

```python
df = pd.read_csv(os.path.join(path, 'PCA5_2010s.csv'))
```

```python
# Libraries for ML
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

## 2. Single Dendrogram

```python
df.columns.tolist
```

```python
df = df.drop(columns = ['Unnamed: 0'])
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Compute the linkage matrix
distance_single = linkage(df.loc[:,['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5']], method="single")

# Plot the dendrogram
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)  # Pass the linkage matrix here
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Single Method PCA5",fontsize=18)
plt.show()
```

## 3. Complete Dendrogram

```python
# Compute the linkage matrix
distance_complete = linkage(df.loc[:,['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5']], method="complete")

# Plot the dendrogram
plt.figure(figsize=(18,6))
dendrogram(distance_complete, leaf_rotation=90)  # Pass the linkage matrix here
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Complete Method PCA5",fontsize=18)
plt.show()
```

## 4. Average Dendrogram

```python
# Compute the linkage matrix
distance_average = linkage(df.loc[:,['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5']], method="average")

# Plot the dendrogram
plt.figure(figsize=(18,6))
dendrogram(distance_average, leaf_rotation=90)  # Pass the linkage matrix here
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Average Method PCA5",fontsize=18)
plt.show()
```

## 5. Ward Dendrogram

```python
# Compute the linkage matrix
distance_ward = linkage(df.loc[:,['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5']], method="ward")

# Plot the dendrogram
plt.figure(figsize=(18,6))
dendrogram(distance_ward, leaf_rotation=90)  # Pass the linkage matrix here
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Ward Method PCA5",fontsize=18)
plt.show()
```
